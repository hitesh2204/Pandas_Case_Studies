### How many people have heart disease and how many people don't have heart disease.

```python
heart.columns
```

```python
heart_disease=heart['target'].value_counts().reset_index()
heart_disease
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='index',y='target',data=heart_disease)
```

### Find the count of male and female in this dataset.

```python
heart.columns
```

```python
sex=heart['sex'].value_counts().reset_index()
sex
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='index',y='sex',data=sex)
plt.xticks([0,1],['Female','male'])
plt.show()
```

### Find the gender distribution according to target columns.

```python
sex_dist=heart.groupby('sex')['target'].value_counts().unstack().unstack()
sex_dist
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='sex',hue='target',data=heart)
plt.xticks([0,1],['Female','male'])
plt.legend(labels=['No Disease','Disease'])
plt.show()
```

### Check the distribution of Age in dataset.

```python
plt.figure(figsize=(12,2))
sns.displot(x='age',data=heart,kde=True)
```

### fidn the most common chect pain type in people.
0 - typical angina
1 - atypical angina
2 - non-anginal pain
3 - asymptomatic

```python
cp=heart['cp'].value_counts().reset_index()
cp
```

```python
plt.figure(figsize=(8,3))
sns.barplot(x='index',y='cp',data=cp)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.show()
```

```python
heart.groupby('cp')['target'].value_counts().unstack()
```

### Show the chest pain distibution as per target variable.

```python
plt.figure(figsize=(8,3))
sns.countplot(x='cp',hue='target',data=heart)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.legend(labels=['No_disease','disease'])
plt.show()
```
