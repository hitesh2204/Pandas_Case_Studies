### In which year there is highest average votting.

```python
movie.columns
```

```python
movie_year=movie.groupby('Year')['Votes'].mean().sort_values(ascending=False)
movie_year=movie_year.reset_index()
print(movie_year)
```

```python
sns.barplot(x='Year',y='Votes',data=movie_year)
```

### In which year there is highest average Revenue .

```python
movie.columns
```

```python
movie_revenue=movie.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False).reset_index()
print(movie_revenue)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Year',y='Revenue (Millions)',data=movie_revenue)
plt.title("Highest avg Revenue by each Year")
plt.ylabel("Highest avg Revenue")
plt.xlabel("Year")
plt.show()
```

### Find the average rating of each director.

```python
movie.columns
```

```python
movie_director=movie.groupby('Director')['Rating'].mean().sort_values(ascending=False).reset_index().head(15)
print(movie_director)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Director',y='Rating',data=movie_director)
plt.title("avg rating by each Director")
plt.ylabel("Average Rating")
plt.xlabel("Director")
plt.xticks(rotation = 85)
plt.show()
```

### Display top 10 lengthy movies title and Runtime.

```python
movie.columns
```

```python
movie_runtime=movie.sort_values('Runtime (Minutes)',ascending=False)[['Title','Runtime (Minutes)']].head(10)
print(movie_runtime)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Title',y='Runtime (Minutes)',data=movie_runtime)
plt.title("Top 10 movie with lengthy runtim")
plt.ylabel("Movie RUntime")
plt.xlabel("Movie")
plt.xticks(rotation = 80)
plt.show()
```

### Display number of movie per year.

```python
movie.columns
```

```python
movie_count=movie.groupby('Year')['Title'].count().sort_values(ascending=False).reset_index()
print(movie_count)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Year',y='Title',data=movie_count)
plt.title("Number of movies per year")
plt.ylabel("Number of movies")
plt.xlabel("Year")
plt.xticks(rotation = 80)
plt.show()
```
